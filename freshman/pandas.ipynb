## Columns

```python
# Set columns
iris.columns = ['A', 'B', 'C', 'D', 'class'] 
iris
```

```python
# change column order
iris[['A', 'C', 'D', 'B', 'class']]
```

```python
# select by column
iris['A']
```

```python
# Select by index
iris.loc[[1, 11, 21]]
```

```python
# Select by column and index
iris.loc[[1, 11, 21], ['A', 'B']]
```

```python
# Select by location
iris.iloc[0:10]
```

```python
# Select by location
iris.iloc[0:10, 0:2]
```

```python
# Boolean
iris['A'] > 7
```

```python
# Select by value
iris[iris['A'] > 7]
```

```python
# Random sampling
sample = iris.sample(n=10)
sample
```

```python
# Random sampling by fraction
iris.sample(frac=0.05)
```

## Feed to function

```python
# assign
sample.assign(a_ten_times=sample['A']*10)
```

```python
# apply--feed a column or row 
sample.iloc[:, 0:4].apply(sum)
```

```python
# apply--feed a column or row 
sample.iloc[:, 0:4].apply(sum, axis=1)
```

```python
# self defined function
def text_function(row):
    return 'class: {class}, A:{A}'.format(**row.to_dict())
    
sample.apply(text_function, axis=1)
```

```python
# self defined function with arguments
def text_function(row, head):
    return head + 'class: {class}, A:{A}'.format(**row.to_dict())
    
sample.apply(text_function, args=('>>> ',),  axis=1)
```

```python
# applymap--element wise function 
sample.iloc[:, 0:4].applymap(lambda x: x/100)
```

```python
# Itterate over rows
for row in iris.itertuples():
    print(row)
```
